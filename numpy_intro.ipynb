#Numpy-Introduction

###Creating Matrix

```python
import numpy as np
arr=np.array([[12,87,45,33],
[90,14,62,77],
[8,59,26,41],
[73,11,95,64],
[50,38,70,29]])
print("The Matrix is \n",arr,"\n")
```

###Creating Transpose Of The Matrix

```python
transposeArr=np.transpose(arr)
print("The Transpose of Matrix is \n",transposeArr,"\n")
```

###Dot Product Of Matrix

```python
dotproduct=np.dot(arr,transposeArr)
print("The Dot Product Of Matrix With Its Transpose is\n",dotproduct,"\n")
```

###Mean Of Each Column

```python
columnwisemean=np.mean(transposeArr,axis=0)
print("The columnwise mean : ",columnwisemean,"\n")
```

###Standard Deviation Of Each Row

```python
rowwisestandarddev=np.std(transposeArr,axis=1)
print("The rowwise standard deviation : ",rowwisestandarddev)
```
