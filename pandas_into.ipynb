Import and Load Data

```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Task2/train.csv")
print(df)
```

Finding count of null values

```python
df.isnull().sum()
```

Handle missing values

```python
age_median=df["Age"].median()
cabin_mode=df["Cabin"].mode()
embarked_mode=df["Embarked"].mode()
df=df.fillna({"Age":age_median,"Cabin":cabin_mode[0],"Embarked":embarked_mode[0]})
```

Grouping the data

```python
group=df.groupby("Sex")
ageMean_Sex=group["Age"].mean()
```

Display Summary

```python
print("\nFirst five rows: \n",df.head(5))
print("\nSummary Of Statistics:\n",df.describe())
print("\nMean Age Grouped By Sex: \n",ageMean_Sex)
```
